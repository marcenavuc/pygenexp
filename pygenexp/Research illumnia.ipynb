```python
import pandas as pd
import csv
```

```python
def read_illumnia(filename):
    return pd.read_csv(filename, sep="\t")

def read_bgx(filename):
    return pd.read_csv(filename, sep="\t", skiprows=8)

def research_illumnia(df, bgx):
    df.columns = ["Probe_Id"] + list(df.columns[1:])
    cols = list(filter(lambda x: not x.startswith("Detect"), df.columns))
    df = df[cols]
    cols = list(df.columns)[1:]
    cols = list(map(lambda s: s.split("_")[0], cols))
    return pd.merge(df, bgx)
```

```python
research_illumnia(read_illumnia("./Downloads/GSE40561_non-normalized/GSE40561_non-normalized.txt"),
                 read_bgx("./Downloads/GSE40561_RAW/GPL6947_HumanHT-12_V3_0_R1_11283641_A/GPL6947_HumanHT-12_V3_0_R1_11283641_A.bgx"))
```
